```
import numpy as np
import matplotlib.pyplot as p
import pandas as pd
print("Premier League Strikers (Goals vs Appearances)")
dataset = pd.read_csv('Dataset.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/2, random_state = 0)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
p.scatter(X_train, y_train, color = 'red', marker='*')
p.plot(X_train, regressor.predict(X_train), color = 'black')
p.title('Training set')
p.xlabel('Appearances')
p.ylabel('Goals')
p.show()
print("\n\n\n")
p.scatter(X_test, y_test, color = 'red', marker='*')
p.plot(X_train, regressor.predict(X_train), color = 'black')
p.title('Test set')
p.xlabel('Appearances')
p.ylabel('Goals')
p.show()
```
